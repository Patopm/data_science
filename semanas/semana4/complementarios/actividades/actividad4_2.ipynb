# Actividad 4.2: Imputación de Datos
**Objetivo:** Practicar técnicas de imputación (Media, Mediana, Moda, Forward/Backward).

```python
import pandas as pd
import numpy as np

# Create dataset with nulls
df = pd.DataFrame({
    "score": [85, np.nan, 90, np.nan, 88, 92, 85],
    "category": ["A", "B", "A", np.nan, "B", "A", "A"]
})
```

### Aplicación de Técnicas de Imputación

```python
# Mean imputation (Best for normal distributions without extreme outliers)
mean_val = df["score"].mean()
df["score_mean"] = df["score"].fillna(mean_val)

# Median imputation (Best when data has outliers)
median_val = df["score"].median()
df["score_median"] = df["score"].fillna(median_val)

# Mode imputation (Best for categorical data)
mode_val = df["category"].mode()[0]
df["category_mode"] = df["category"].fillna(mode_val)

# Forward and Backward fill (Best for time-series data)
df["score_ffill"] = df["score"].ffill()
df["score_bfill"] = df["score"].bfill()

print(df)
```

### Justificación
- **Media**: Útil para variables continuas con distribución simétrica.
- **Mediana**: Ideal para variables continuas asimétricas o con outliers.
- **Moda**: La mejor opción para variables categóricas.
- **Forward/Backward Fill**: Excelente para series temporales donde el valor actual está correlacionado con el anterior/siguiente.
