# Actividad 4.1: Identificación y Manejo de Valores Faltantes
**Objetivo:** Aprender a detectar y manejar valores faltantes en un DataFrame.

```python
import pandas as pd
import numpy as np

# Create a DataFrame with missing values
data = {
    "id": [1, 2, 3, 4, 5],
    "name": ["Alice", "Bob", np.nan, "David", "Eve"],
    "age": [25, np.nan, 30, 22, np.nan],
    "salary": [50000, 60000, 55000, np.nan, 65000]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

### Detección de Valores Faltantes
Uso de `isnull()`, `notnull()`, `sum()` e `info()`.

```python
# Check for null values
print("\n--- isnull() ---")
print(df.isnull())

print("\n--- sum() of nulls per column ---")
print(df.isnull().sum())

print("\n--- info() ---")
df.info()
```

### Estrategias de Manejo
Aplicaremos 3 técnicas:
1. Eliminar filas con nulos.
2. Rellenar con un valor constante.
3. Rellenar usando el valor anterior (forward fill).

```python
# Technique 1: Drop rows with any null values
df_dropped = df.dropna()
print("\nDropped nulls:\n", df_dropped)

# Technique 2: Fill with constant value
df_constant = df.fillna({"name": "Unknown", "age": 0, "salary": 0})
print("\nFilled with constant:\n", df_constant)

# Technique 3: Forward fill
df_ffill = df.ffill()
print("\nForward fill:\n", df_ffill)
```
