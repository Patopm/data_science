# Actividad 4.3: Transformación de Datos
**Objetivo:** Normalización, estandarización, One-Hot Encoding y variables derivadas.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

df = pd.DataFrame({
    "price": [10, 20, 30, 40, 50],
    "weight": [100, 150, 200, 120, 300],
    "color": ["red", "blue", "red", "green", "blue"]
})

display(df)
```

### Normalización y Estandarización

```python
# Min-Max Normalization
scaler_minmax = MinMaxScaler()
df[["price_norm", "weight_norm"]] = scaler_minmax.fit_transform(
    df[["price", "weight"]]
)

# Standardization (Z-score)
scaler_std = StandardScaler()
df[["price_std", "weight_std"]] = scaler_std.fit_transform(
    df[["price", "weight"]]
)

display(df)
```

### Codificación Categórica y Variables Derivadas

```python
# One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=["color"], prefix="color")

# Derived Variable: price per weight
df_encoded["price_per_weight"] = (
    df_encoded["price"] / df_encoded["weight"]
)

display(df_encoded.head())
```

### Explicación
- **Min-Max**: Útil cuando el algoritmo no asume ninguna distribución y se requiere que los datos estén en una escala limitada (ej. Redes Neuronales).
- **Z-score**: Útil cuando los datos tienen una distribución gaussiana (ej. SVM, Regresión Logística).
- **One-Hot Encoding**: Necesario para que los modelos procesen texto/categorías sin asumir orden jerárquico.
- **Variables derivadas**: Capturan información combinada que ayuda al modelo a encontrar patrones más fácilmente.
